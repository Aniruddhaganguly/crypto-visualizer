```python
#Importing Packages
import requests
import pandas as pd
from datetime import datetime
```

```python
# fucntion to get data for etherium function
def get_eth_market_chart(vs_currency='usd', days='30'):
    url = f'https://api.coingecko.com/api/v3/coins/ethereum/market_chart'
    params = {'vs_currency': vs_currency, 'days': days}
    response = requests.get(url, params=params)
    data = response.json()
    
    prices = data['prices']
    df = pd.DataFrame(prices, columns=['timestamp', 'price'])
    df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
    return df
```

```python
#Etherium data
eth_data=get_eth_market_chart()
eth_data.head()
```

```python
def convert_to_ohlc(df):
    df.set_index('date', inplace=True)
    ohlc = df['price'].resample('1D').ohlc()
    return ohlc.dropna()
```

```python
ohlc=convert_to_ohlc(eth_data)
ohlc.head()
```

```python
#plotting the candlesticks for 30 days data of etherium

import webbrowser

def plot_candlestick_to_browser(ohlc, filename='eth_candlestick.html'):
    import plotly.graph_objects as go
    
    fig = go.Figure(data=[go.Candlestick(
        x=ohlc.index,
        open=ohlc['open'],
        high=ohlc['high'],
        low=ohlc['low'],
        close=ohlc['close'],
        increasing_line_color='green',
        decreasing_line_color='red'
    )])
    
    fig.update_layout(
        title='Ethereum Price Candlestick Chart (30 Days)',
        xaxis_title='Date',
        yaxis_title='Price (USD)',
        xaxis_rangeslider_visible=False
    )
    
    # Save to HTML
    fig.write_html(filename)
    
    # Open in default web browser
    webbrowser.open_new_tab(filename)


plot_candlestick_to_browser(ohlc)


```
